## Speed distribution function graph

Serway, Physics chapter 21-5

[Fundamental Physical Constants from NIST](https://physics.nist.gov/cuu/Constants/)

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
N=1e5    # number of molecules
amu=1.660539e-27   # atomic mass unit, kg
m=28*amu           # mass of nitrogen molecule, kg
k=1.380649e-23     # Boltzmann constant, J/K
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.set_title('speed distribution function')
ax.set_xlabel('v(m/s)')
ax.set_ylabel('Nv[molecules/(m/s)]')

ax.set_xlim(0, 1600)
ax.set_ylim(0, 200)

v=np.linspace(0,1600,100)

for j in range(1,4):
    T=300*j
    b=m/(2*k*T)
    Nv=4*pi*N*((b/pi)**1.5)*(v**2)*np.exp(-b*v**2)
    ax.plot(v, Nv, label=f'{T} K')

#plt.grid()
ax.legend()
plt.show()    
```
